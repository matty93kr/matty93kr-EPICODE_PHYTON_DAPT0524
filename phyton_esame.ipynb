```python
#Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati:
# a. l'evoluzione del casi totali dall'inizio alla fine dell'anno 
```

```python
covid['date'] = pd.to_datetime(covid['date'])
italia_2022 = covid[(covid["location"] == "Italy") & (covid["date"].dt.year == 2022)]
```

```python
italia_2022
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
plt.plot(italia_2022["date"], italia_2022["total_cases"], color = "green", label = "Casi totali")
plt.title("evoluzione dei casi totali in italia nel 2022")
plt.xlabel("Data")
plt.ylabel("casi totali")
plt.grid(True)
plt.show

```

```python
# b.il numero di nuovi casi rispetto alla data
```

```python
italia_2022.loc[:, 'new_cases'] = italia_2022['total_cases'].diff().fillna(0)
```

```python
print(italia_2022)
```

```python
sns.lineplot(data=italia_2022, x="date", y="new_cases", color="blue", marker='o', label="Nuovi Casi")
plt.xlabel("Data",)
plt.ylabel("Nuovi Casi",)
plt.grid
plt.show
```

```python
# k. poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni
```

```python
filtro_ita22 = italia_2022[(italia_2022["new_cases"].notna()) & (italia_2022["new_cases"] > 0)]
print(filtro_ita22)
```

```python
# punto 4 Riguardo le nazioni di Italia, Germania e Francia 
# mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) 
#da maggio 2022 (incluso) ad aprile 2023 (incluso
```

```python
covid["date"] = pd.to_datetime(covid["date"])
```

```python
inizio_data = "2022-05-01"
fine_data = "2023-04-30"
nazioni = ["Italy", "Germany", "France"]
```

```python
covid_filtro = covid[
    (covid["date"] >= inizio_data) &
    (covid["date"] <= fine_data) &
    (covid["location"].isin(nazioni))]
```

```python
covid_filtro
```

```python
covid_filtro.boxplot(column="icu_patients", by="location",grid=False)
plt.title("Pazienti in terapia in tesniva ( Maggio 2022 - Aprile 2023)")
plt.xlabel("nazioni")
plt.ylabel("pazienti in terapia intensiva")
plt.show
          
```
